```python
import pandas as pd
import json

def csv_to_json_bulk(csv_file_path, json_file_path):
    json_bulk = ""

    # CSV 파일 읽기
    df = pd.read_csv(csv_file_path, encoding="utf-8-sig")

    # 각 행에 대해 처리
    for index, row in df.iterrows():
        # JSON 형식으로 변환
        json_data = row.to_json(orient='records', force_ascii=False)[1:-1]

        # Bulk JSON 형식으로 포맷팅
        json_bulk += f'{{"index": {{"_index": "learning", "_id": "{row["id"]}"}}}}\n'
        json_bulk += "{" + "\"title\" : \"" + row["title"] + "\"}\n"
        # json_bulk_list.append(json_bulk)

    print(json_bulk)

    # JSON 파일에 쓰기
    with open(json_file_path, 'w', encoding='utf-8') as json_file:
        json_file.write(json_bulk)

# CSV 파일 경로
csv_file_path = 'learning.csv'
json_file_path = 'learning.json'

# CSV 파일을 읽고 Bulk JSON 형식으로 변환
csv_to_json_bulk(csv_file_path, json_file_path)
```
